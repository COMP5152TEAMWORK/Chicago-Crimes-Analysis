
# This is the study of the relationship between the incidence of crime (both total number and type-specific number) and the time (date and time stamp), for this purpose, the original files undergo the following steps:
### 1. Remove NA
### 2. Among the original features, selected ID, Day & Time stamp (splitting original Date),
### Primary Type, Year 
### 3. A file concatenating all the years' data is generated (df_ALL)



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
style.use('ggplot')


DATA_DIR = '../data/'
DATA_GEN_DIR = '../data_gen/'
tag2 = 'AllYearData_NAremoved'
suffix = '.csv'
years = list(range(2010,2023))
print(years)

df_ALL = pd.DataFrame() # Initialize an empty dataframe, for later usage of concatenation all year's df
df_ALL['ID'] = []
df_ALL['Day'] = []
df_ALL['Time'] = []
df_ALL['Primary Type'] = []
df_ALL['Year'] = []


for y in years:
    df = pd.read_csv('{}{}{}'.format(DATA_DIR, y, suffix))
    print(f'-------NA value summary of the year {y} BEFORE drop NA-------' )
    print(df.isnull().sum())
    df = df.dropna(how = 'any')
    print(f'-------NA value summary of the year {y} AFTER drop NA-------' )
    print(df.isnull().sum())
    df_select = df[['ID','Date','Primary Type','Year']]
    df_select['Day'] = df_select['Date'].str[0:10]
    df_select['Time'] = df_select['Date'].str[11:22]
    df_select = df_select[['ID','Day','Time','Primary Type','Year']]
    df_ALL = pd.concat([df_ALL, df_select])
    
df_ALL.to_csv('{}{}{}'.format(DATA_GEN_DIR,tag2,suffix), index=False)

```

# Time_Crime relationship across a day (which time slot see most crimes)

# Define a funciton to convert the Time into 24hour format

```python
def timeConversion(time):
    if time[-2:] == 'AM' and time[0:2]=='12':
        return "00" + time[2:-3]
    elif time[-2:]=='AM':
        return time[:-3]
    elif time[-2:] =='PM' and time[0:2] == '12':
        return time[:-3]
    else:
        return str(int(time[:2]) + 12)+ time[2:8]
    

    
    
    
```

# Extract the hour information from the Time, count the number of crimes group by hour and plot the total number across 24 hours

```python
df_ALL["Time2"] = df_ALL["Time"].apply(timeConversion)
df_ALL = df_ALL.sort_values('Time2',ascending = True)
df_ALL["Time3"] = pd.to_datetime(df_ALL['Time2'], format='%H:%M:%S')
df_ALL["Hour"] = df_ALL["Time3"].dt.hour
df_ALL.tail()

df_ALL.groupby("Hour",as_index=False).count()

df = df_ALL.groupby('Hour', as_index=False)['ID'].count().sort_values('Hour',ascending = True)
plt.plot(df['Hour'], df['ID'])
plt.xticks(rotation = 90)
plt.title("Time-Crime dependency: What time is most dangerous (summed all between 2010 - 2022)")
plt.ylim(0,250000)
plt.xticks(np.arange(0,25,2))

```
